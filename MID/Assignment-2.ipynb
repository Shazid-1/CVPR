# Activation Functions

```python
import matplotlib.pyplot as plt
import numpy as np
  
x = np.arange(-10, 11, 1)
print(x)

def plot_graph(y,ylabel):
    plt.figure(figsize=(5,3))
    plt.plot(x,y, 'o--')
    plt.grid(True)
    plt.xlabel("X")
    plt.ylabel(ylabel)
    plt.show()
```

### Step function

```python
y = list(map(lambda n: 1 if n>0.5 else 0, x))
plot_graph(y,"STEP(X)")
```

Advantages:
+ It is effortless to understand and implement.
+ We can do a beneficial binary classification task where it takes 1/0 as input and gives output as 1/0.
+ We can use this function in a neural network as a nonlinear function.

Disadvantages:
+ The step function produces a binary output (0 or 1) regardless of the input value. As a result, optimizing the neural network to get reliable results is challenging.
+ This function is not 0-centred; many neural network topologies depend on the output values centred around 0. Therefore, it might be problematic for some of them.
+ The gradient information is lost in that range since the step function provides a constant output(1/0) for a range of input values.

### Sigmoid function

```python
y = 1 / (1 + np.exp(-x))
plot_graph(y, "Sigmoid(X)")
```

Advantages :
+ It is suitable for use in binary classification problems at the output layer of neural networks since the sigmoid function's value lies between 0 and 1, and may predict output by maintaining a threshold value between 0 and 1.
+ It has a smooth gradient, preventing jumps in output values.

Disadvantages :
- It is Computationally expensive
- Vanishing gradient: for very high or very low values of X, the prediction has almost no change, causing a vanishing gradient problem. It can result in the network needing to learn further or being too slow to reach an accurate prediction.
- Its outputs are not zero-centred

### Tanh function

```python
y = (np.exp(2*x) - 1) / (np.exp(2*x) + 1)
plot_graph(y, "TanH(X)")
```

Advantages :
+ It is usually used in hidden layers of neural networks and gives values between -1 and 1. It will make mean for the hidden layer 0 or very close to itself and help the centring of data and learning for the next layer much more accessible.
+ Effective for modelling extremely positive, neutral, and strongly negative inputs.

Disadvantages : 
+ The prediction for very high or shallow values of x remains the same, causing a vanishing gradient problem. Due to this, a neural network may refuse to learn further, or it will become too slow to find the accurate output.
+ It is also computationally expensive.


### Relu function

```python
y = list(map(lambda a: a if a>=0 else 0, x))
plot_graph(y,"ReLU(X)")
```

Advantages:
+ The Cost Of Computation Is Less As Compared To Sigmoid & Tan H Since It Uses Simple Mathematical Computations. This Function Activates Only A Few Neurons At A Time Which Makes The Neural Network Sparse & Computationally Efficient.
+ It Is Faster Than The Sigmoid & Tan H Activation Function.
+ The Derivative Function In ReLU Allows Backpropagation.

Disadvantages:
- The Network Cannot Perform Backpropagation When The Gradient Of The Function Becomes Zero Due To the Zero Or Negative Value Of The Input.

### Elu function

```python
alpha = 1.0
y = np.where(x>=0, x, alpha*(np.exp(x)-1))
plot_graph(y, "ELU(x)")
```

Advantages:
- It has all the benefits of the ReLU function.
- Closer to zero mean outputs

Disadvantages:
- A negative saturation regime compared with Leaky ReLU adds some robustness to noise.
- Not all varieties of neural networks can utilize the ELU function.
- The ELU function is more computationally expensive than ReLU.

### Selu function

```python
alpha=1.6733
scale=1.0507
y = np.where(x > 0, scale*x, scale*(alpha*np.exp(x)-alpha))

plot_graph(y, "SELU(x)")
```

Advantages :
- This function can self-normalize. Therefore, the function's output is nearly normally distributed for a given input with a zero mean and a single variance.
- The gradients are kept from getting too tiny or too big by the function, which is built to maintain each neuron's output within a certain range.
- Automatically computes the values of lambda and alpha during training


Disadvantages : 
- The SELU function is only applicable to feedforward neural networks with dense layers and may not work well for other types.
- The SELU function is sensitive to weight initialization.
